```python
import pandas as pd
import os
import requests
```

```python
def get_url_file(url, local_dir="./temp"):
    response = requests.get(url)
    if response.status_code == 200:
        file_name = url.split("/")[-1]
        file_path = os.path.join(local_dir, file_name)
        with open(file_path, "wb") as f:
            f.write(response.content)
        return file_path
    else:
        return None
```

```python
from interface import AudioDectector

audio_detector = AudioDectector()
```

```python
def judge(df_data):
    urls = df_data["素材下载地址"]
    genre = []
    emotion = []
    instruments = []
    gender = []
    for i in range(len(urls)):
        url = urls[i]
        temp_file_path = get_url_file(url)
        if temp_file_path:
            res = audio_detector.detect(temp_file_path)
            genre.append(res["genre"])
            emotion.append(res["emotion"])
            instruments.append(res["instruments"])
            gender.append(res["gender"])
        else:
            genre.append("error")
            emotion.append("error")
            instruments.append("error")
            gender.append("error")

    df_data["风格"] = genre
    df_data["情感"] = emotion
    df_data["乐器"] = instruments
    df_data["性别"] = gender
    return df_data
```
